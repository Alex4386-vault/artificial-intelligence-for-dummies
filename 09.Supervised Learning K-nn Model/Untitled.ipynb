```python
from sklearn import datasets
import pandas as pd
```

## Learn with Scikit-Learn

```python
import pandas as pd
import numpy as np
from sklearn import neighbors
```

### Load K-nn Classifier Model

```python
knn = neighbors.KNeighborsClassifier()
```

### Training

```python
knn.fit(iris['data'], iris['target'])
```

### Training Evaluation

```python
knn.score(iris['data'], iris['target'])
```
